# 1. 간단한 크롤링

```python
import requests
from bs4 import BeautifulSoup
```

```python
hani_URL = "http://www.hani.co.kr/arti/society/home01.html" #article_popularity > list > a
```

```python
chosun_URL = "http://news.chosun.com/politics/" #.tab_link > a 
```

```python
headers = {"Accept-Encoding":"false"}
```

```python
def news_scrap(URL):
    headers = {"Accept-Encoding":"false"}
    html = requests.get(URL, headers = headers)
    URLStr = html.content.decode('utf-8')
    return BeautifulSoup(URLStr, "lxml")
```

```python
soup1 = news_scrap(hani_URL)
```

```python
soup1
```

```python
for soup in soup1.select(".article-area"):
    print(soup.get_text().strip())
```

## 3. 셀레니움
1. 셀레니움으로 들어간다. 
2. 태그를 찾아서 그 안으로 들어간다. 
3. 내용을 크롤링 한 다음에 저장한다. 

```python
def doSelenium(URL, params):
    from selenium import webdriver
    driver = webdriver.Chrome()
    html = requests.get(URL, params=params)
    driver.get(html.url)
    tagList = getMacro(html.url)
    for i in range(len(tagList)):
        newURL = tagList[i].get('href')
        dom = news_scrap1(newURL)
        print(dom.select(".text")[1].get_text())
```

```python
def doSelenium2(URL, params):
    from selenium import webdriver
    driver = webdriver.Chrome()
    html = requests.get(URL, params=params)
    driver.get(html.url)
    tagList = getMacro(html.url)
    newURL = tagList[1].get('href')
    dom = news_scrap1(newURL)
    print(dom.select(".text")[1].get_text())
```
